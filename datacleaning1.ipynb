```python
import pandas as pd
df1 = pd.read_csv("C:/DATASCIENCE/test1.csv")
df2 = pd.read_csv("C:/DATASCIENCE/train1.csv")
print("Original Data Shape:", df1.shape,df2.shape)
```

```python
# Dataset 1 info
print("Dataset 1:")
print(df1.info())
print(df1.describe())
print(df1.isnull().sum())
```

```python
 #Dataset 2 info
print("\nDataset 2:")
print(df2.info())
print(df2.describe())
print(df2.isnull().sum())
```

```python

print(df1.isnull().sum())

print(df2.isnull().sum())
```

```python
# # Fill all numeric NaNs with mean, and categorical with mode
for column in df1.columns:
     if df1[column].dtype == 'object':
         df1[column] = df1[column].fillna(df1[column].mode()[0])
     else:
         df1[column] = df1[column].fillna(df1[column].mean())

for column in df2.columns:
     if df2[column].dtype == 'object':
         df2[column] = df2[column].fillna(df2[column].mode()[0])
     else:
         df2[column] = df2[column].fillna(df2[column].mean())

```

```python
print("Missing values in df1:\n", df1.isnull().sum())
print("\nMissing values in df2:\n", df2.isnull().sum())
```

```python
# Convert 'Dependents' like '3+' or '2+' to numeric
df1['Dependents'] = df1['Dependents'].replace({'3+': 3, '2+': 2})

df2['Dependents'] = df2['Dependents'].replace({'3+': 3, '2+': 2})
print(df1['Dependents'].head(20))
print(df2['Dependents'].head(20))
```

```python
print(df1['Credit_History'])
```

```python
df1['Credit_History'] = df1['Credit_History'].fillna(df1['Credit_History'].mode()[0])

df1['Credit_History'] = df1['Credit_History'].astype(int)
df2['Credit_History'] = df2['Credit_History'].fillna(df2['Credit_History'].mode()[0])

df2['Credit_History'] = df2['Credit_History'].astype(int)
```

```python
print(df1['Credit_History'])
print(df2['Credit_History'])
```
